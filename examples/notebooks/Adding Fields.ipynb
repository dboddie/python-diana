```python
import datetimefrom metno import bdiana, plotcommandsfrom metno.ipython_extensions import embed
```

```python
b = bdiana.BDiana()b.setup()
```

```python
b.addModel("baltic_sea_ice_thickness", "http://thredds.met.no/thredds/dodsC/myocean/siw-tac/siw-fmi-baltic/2013/03/ice_thic_baltic_201303311400.nc", clear = True)
```

```python
models = list(b.getModels())models.sort()models
```

```python
fields = list(b.getFields(models[0]))fields.sort()fields
```

```python
a = plotcommands.Area(name = "Europa")m = plotcommands.Map(map = "Gshhs-Auto", backcolour = "white", land = "on")m.setOption("land.colour", "flesh")f = plotcommands.Field(model = models[0], plot = fields[1], plottype = "fill_cell")times = b.getFieldTimes(models[0], fields[1])times
```

```python
b.setPlotCommands([a.text(), m.text(), f.text()])if times:    b.setPlotTime(times[-1])else:    b.setPlotTime(datetime.datetime.now())
```

```python
im = b.plotImage(500, 500)
```

```python
embed(im)
```
